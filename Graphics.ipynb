```python
%matplotlib inline

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

## Graph of evolution for single game

```python
84, 111, 231

pd1 = pd.read_csv("./results_Boxing-v0_84_oCV.csv")
pd2 = pd.read_csv("./results_Boxing-v0_111_oCV.csv")
pd3 = pd.read_csv("./results_Boxing-v0_231_oCV.csv")

a = pd.concat([pd1, pd2, pd3])

# a = pd.read_csv("./results_MsPacman-v0_231_oCV.csv")

tt = a.replace([np.inf, -np.inf], np.nan).dropna()
a = tt.groupby("Generation")
tt
```

```python
stats = pd.DataFrame()
stats["Maximum"] = a.max()["Fitness"]
stats["Minimum"] = a.min()["Fitness"]
stats["Average"] = a.mean()["Fitness"]
```

```python
stats.plot()
```

```python
plt.subplots(figsize=(27, 7))

sns.boxplot(x="Generation", y="Fitness", data=tt)
```
